```python
import pandas as pd
```

```python
df = pd.read_parquet('../../ICU/tables/microbiologyevents.parquet')
#df['test_name_org_name'] = df['test_name'].astype(str)+ '_' + df['org_name'].astype(str)
df
```

```python
df['subject_id_hadm_id'] = df['subject_id'].astype(str) + '_' + df['hadm_id'].astype(str)
df
```

```python
df["subject_id_hadm_id"].nunique()
```

```python
top_10 = df['test_name'].value_counts().head(50).index.tolist()
df = df[df['test_name'].isin(top_10)]
#df = df.drop_duplicates(subset=['subject_id', 'hadm_id', 'test_name'])
df['subject_id_hadm_id'] = df['subject_id'].astype(str) + '_' + df['hadm_id'].astype(str)
#df_micro = df_micro[['test_name','label']]
#merged_lab_events_df = pd.merge(df_labevents, df_lab_items, on='itemid', how='left')
df_micro1 = df.pivot_table(
    index=['subject_id', 'hadm_id', 'subject_id_hadm_id'], 
    columns='test_name', 
    values='org_name', 
    aggfunc='first'
)
```

```python
df_micro1
```

```python
df_micro1=df_micro1.drop_duplicates()
df_micro1
```

```python
df_micro1=df_micro1.reset_index()
df_micro1
```

```python
top_10 = df['ab_name'].value_counts().head(50).index.tolist()
df = df[df['ab_name'].isin(top_10)]
df = df.drop_duplicates(subset=['subject_id', 'hadm_id', 'ab_name'])
df['subject_id_hadm_id'] = df['subject_id'].astype(str) + '_' + df['hadm_id'].astype(str)
#df_micro = df_micro[['test_name','label']]
#merged_lab_events_df = pd.merge(df_labevents, df_lab_items, on='itemid', how='left')
df_micro2 = df.pivot_table(
    index=['subject_id', 'hadm_id', 'subject_id_hadm_id'], 
    columns='ab_name', 
    values='interpretation', 
    aggfunc='first'
)
```

```python
df_micro2
```

```python
df_micro2=df_micro2.drop_duplicates()
df_micro2
```

```python
df_micro2=df_micro2.reset_index()
df_micro2
```

```python
df_micro2["subject_id_hadm_id"].nunique()
```

```python
#Remove rows with missing values
df_micro1=df_micro1.dropna(how='all')
df_micro2=df_micro2.dropna(how='all')
```

```python
df_micro1=df_micro1.drop(["subject_id", "hadm_id"], axis=1)
df_micro2=df_micro2.drop(["subject_id", "hadm_id"], axis=1)
```

```python
merged_micro = pd.merge(df_micro1, df_micro2, on='subject_id_hadm_id', how='inner')
merged_micro
```
