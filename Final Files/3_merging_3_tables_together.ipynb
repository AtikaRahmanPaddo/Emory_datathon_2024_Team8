```python
import pandas as pd
```

```python
merge_1=pd.read_csv("merged_admissions_patients.csv")
merge_2=pd.read_csv("df_icustays.csv")
```

```python
merge_1=merge_1.drop(["Unnamed: 0"], axis=1)
merge_2=merge_2.drop(["Unnamed: 0"], axis=1)
```

```python
merge_1=merge_1.drop_duplicates()
merge_2=merge_2.drop_duplicates()
```

```python
merge_1["subject_id_hadm_id"]=merge_1["subject_id"].astype(str)+'_'+merge_1['hadm_id'].astype(str)
```

```python
merge_1=merge_1.drop(["subject_id", "hadm_id"], axis=1)
```

```python
merge_1
```

```python
merged_4 = pd.merge(merge_1, merge_2, on='subject_id_hadm_id', how='inner')
merged_4
```

```python
merged_4.to_csv("merged_4.csv")
```
