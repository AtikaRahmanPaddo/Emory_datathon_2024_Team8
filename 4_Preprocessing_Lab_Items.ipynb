```python
import pandas as pd
```

```python
df_items = pd.read_parquet('../../ICU/tables/d_items.parquet')
```

```python
df_lab_items = pd.read_parquet('../../ICU/tables/d_labitems.parquet')
```

```python
df_labevents = pd.read_parquet('../../ICU/tables/labevents.parquet')
```

```python
top_10 = df_labevents['itemid'].value_counts().head(25).index.tolist()
```

```python
df_labevents = df_labevents[df_labevents['itemid'].isin(top_10)]
```

```python
df_labevents = df_labevents.drop_duplicates(subset=['subject_id', 'hadm_id', 'itemid'])
```

```python
df_labevents['subject_hadm_id'] = df_labevents['subject_id'].astype(str) + '_' + df_labevents['hadm_id'].astype(str)
```

```python
df_lab_items = df_lab_items[['itemid','label']]
```

```python
merged_lab_events_df = pd.merge(df_labevents, df_lab_items, on='itemid', how='left')
```

```python
lab_events_df = merged_lab_events_df.pivot_table(
    index=['subject_id', 'hadm_id', 'subject_hadm_id'], 
    columns='label', 
    values='value', 
    aggfunc='first'
)
```

```python
lab_events_df = lab_events_df.dropna()
lab_events_df = lab_events_df.drop_duplicates()
```

```python
lab_events_df = lab_events_df.reset_index()
lab_events_df = lab_events_df.drop_duplicates(subset=['subject_hadm_id'])
```

```python
lab_events_df
```

```python
lab_events_df.rename(columns={'subject_hadm_id': 'subject_id_hadm_id'}, inplace=True)
```

```python
lab_events_df=lab_events_df.drop(["subject_id", "hadm_id"], axis=1)
```

```python
merged_final = pd.merge(merged_previous, lab_events_df, on='subject_id_hadm_id', how='inner')
merged_final
```
